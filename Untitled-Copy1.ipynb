#### find all the urls of card divs in a page

```python
urls=[]
main_div = edusanjalpage.find('div', class_='grid gap-8 md:grid-cols-2 lg:grid-cols-3 lg:max-w-full')
div_list = main_div.find_all('div', class_='overflow-hidden bg-white rounded shadow-xl')
for div in div_list:
    link = div.find('a')['href']
    print(link)
    link = link[8:]
    total_url= current_website[:-8]+link
    print(total_url)
    urls.append(total_url)
# print(urls)
```

```python
for url in urls:
    response = requests.get(url,headers=HEADER)
    html_content = response.text
    soup = bs(html_content,"lxml")
    print(soup.prettify())
```
